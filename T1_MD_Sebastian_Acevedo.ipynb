#Descripción de la Base de Datos
---
Este estudio tiene como objetivo comprender que factores influyen en el rendimiendo académico de jóvenes entre 15 y 18 años
##Atributos:
- **StudentID:** Identificador único para cada estudiante.
- **Age:** Edad del estudiante.
- **Gender:** El género del estudiante.
  - 0: Hombre
  - 1: Mujer
- **Ethnicity:** Etnia del estudiante.
  - 0: Caucásico
  - 1: Afroamericano
  - 2: Asiático
  - 3: Otro
- **ParentalEducation:** El nivel de educación que tienen los padres.
  - 0: Ninguno
  - 1: Escuela Secundaria
  - 2: Algo de estudios Universitarios
  - 3: Lienciatura
  - 4: Magister o doctorado
- **StudyTimeWeekly:** Tiempos de estudio por semana, en horas.
- **Absences:** Número de ausencias durante el año escolar.
- **Tutoring:** Posee tutor.
  - 0: No
  - 1: Si
- **ParentalSupport:** Nivel de apoyo parental.
  - 0: Ninguno
  - 1: Bajo
  - 2: Moderado
  - 3: Alto
  - 4: Muy alto
- **Extracurricular:** El estudiante participa en actividades extracurriculares.
  - 0: No
  - 1: Si
- **Sports:** El estudiante participa en algún deporte.
  - 0: No
  - 1: Si
- **Music:** El estudiante participa en acrividades relacionadas a la música.
  - 0: No
  - 1: Si
- **Volunteering:** El estudiante participa en algún voluntariado.
  - 0: No
  - 1: Si
- **GPA:** Es el promedio de calificaciones. Va en escala de 2.0 a 4.0
- **GradeClass** Calificación de los estudiantes basado en la columna GPA.
  - 0: 'A' (GPA >= 3.5)
  - 1: 'B' (3.0 <= GPA < 3.5)
  - 2: 'C' (2.5 <= GPA < 3.0)
  - 3: 'D' (2.0 <= GPA < 2.5)
  - 4: 'F' (GPA < 2.0)

##2. ¿Qué impacto tiene el nivel de apoyo parental en la calificación final de los estudiantes?

**Instrucciones:**
- Genera un gráfico de barras que compare el promedio de estudiantes (eje y) según el nivel de apoyo parental (eje x).
- Indica la diferencia porcentual entre los estudiantes sin ningún apoyo y los que reciben un muy alto apoyo.
- Redacta una conclusión.

```python
# Calcular el promedio de GPA por nivel de apoyo parental
promedio_por_apoyo = df.groupby('ParentalSupport')['GPA'].mean()

# Gráfico
plt.figure(figsize=(8,8))
bars = plt.bar(promedio_por_apoyo.index, promedio_por_apoyo.values, color=['skyblue', 'teal', 'skyblue', 'teal', 'skyblue'])
plt.xlabel('Nivel de Apoyo Parental')
plt.ylabel('Promedio de GPA')
plt.title('Promedio de GPA según nivel de apoyo parental')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, f'{yval:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Porcentaje de diferencia
sin_apoyo = promedio_por_apoyo.loc[0]
con_apoyo = promedio_por_apoyo.loc[4]
Porcentaje = ((con_apoyo - sin_apoyo) / sin_apoyo) * 100
print()
print(f"Diferencia porcentual entre apoyo 0 y apoyo 4: {Porcentaje:.2f}%")
```

Con este gráfico queda en evidencia el impacto que tiene el apoyo de los padres con respecto a su desempeño a la hora de obtener calificaciones, ya que a medida de que sube el apoyo parental, sube también el valor del GPA.
Es mas, existe una diferencia de GPA del 42.3% entre los estudiantes sin apoyo parental y los con mucho apoyo parental, significando que a los estudiantes sin apoyo les va un 42% peor que a los que si tienen mucho apoyo.

##11. Determinar si la distribución de GPA sigue una distribución normal en los grupos genero, etnicidad y tutoría.

**Instrucciones:**
- Obten la desviación estandar del 'GPA' de las mujeres.
- Obten la desviación estandar del 'GPA' de los que reciben tutorias.
- Obten la desviación estandar del 'GPA' de los de etnicidad asiática.


- Genera un histograma que denote la distribución del 'GPA' de las mujeres.
- Genera un histograma que denote la distribución del 'GPA' de los que reciben tutorias.
- Genera un histograma que denote la distribución del 'GPA' de los de etnicidad asiática.


- Obten el porcentaje de datos dentro de 1σ (1 * desviación estandar) para los datos de las mujeres.
- Obten el porcentaje de datos dentro de 2σ (2 * desviación_estandar) para los datos de las mujeres.
- Concluye si la distribución de datos  de 'GPA' es normal para las mujeres a través de lo obtenido previamente.

```python
# Mujeres
dsv_mujeres = df[df['Gender'] == 1]['GPA'].std()
print(f"Desviación estándar del GPA (mujeres): {dsv_mujeres:.2f}")

# Con tutorías
dsv_tutoria = df[df['Tutoring'] == 1]['GPA'].std()
print(f"Desviación estándar del GPA (tutoría): {dsv_tutoria:.2f}")

# Etnicidad asiática
dsv_asia = df[df['Ethnicity'] == 2]['GPA'].std()
print(f"Desviación estándar del GPA (asiáticos): {dsv_asia:.2f}")

# Histograma Mujeres
plt.figure(figsize=(6,6))
plt.hist(df[df['Gender'] == 1]['GPA'], bins=15, color='orchid', edgecolor='black')
plt.title('Distribución de GPA - Mujeres')
plt.xlabel('GPA')
plt.ylabel('Frecuencia')
plt.show()

# Histograma Tutoría
plt.figure(figsize=(6,6))
plt.hist(df[df['Tutoring'] == 1]['GPA'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribución de GPA - Con Tutoría')
plt.xlabel('GPA')
plt.ylabel('Frecuencia')
plt.show()

# Histograma Asiáticos
plt.figure(figsize=(6,6))
plt.hist(df[df['Ethnicity'] == 2]['GPA'], bins=15, color='lightgreen', edgecolor='black')
plt.title('Distribución de GPA - Etnicidad Asiática')
plt.xlabel('GPA')
plt.ylabel('Frecuencia')
plt.show()

# GPA de mujeres
gpa_mujeres = df[df['Gender'] == 1]['GPA']
media = gpa_mujeres.mean()
dsv = gpa_mujeres.std()

# 1 desviación estándar
un_sigma = gpa_mujeres[(gpa_mujeres >= media - dsv) & (gpa_mujeres <= media + dsv)]
porcentaje_un_sigma = len(un_sigma) / len(gpa_mujeres) * 100
print(f"Porcentaje dentro de 1σ (mujeres): {porcentaje_un_sigma:.2f}%")

# 2 desviaciones estándar
dos_sigma = gpa_mujeres[(gpa_mujeres >= media - 2*dsv) & (gpa_mujeres <= media + 2*dsv)]
porcentaje_dos_sigma = len(dos_sigma) / len(gpa_mujeres) * 100
print(f"Porcentaje dentro de 2σ (mujeres): {porcentaje_dos_sigma:.2f}%")
```

La distribución de la desviación en las mujeres es bastante simétrica y cercana a la media, con un mayor número de estudiantes concentrados en los rangos de GPA de 1.5 a 2.5. A comparación del resto de gráficos las mujeres presentan mayor frecuencia en todos los niveles, indicando una mayor cantidad de datos.

La distribución de la desviación en estudiantes con tutoría tiene una tendencia similar a la de las mujeres, pero con frecuencias mucho menores. Eso si hay que notar que a pesar de que si bien tambíen hay una concentración en el GPA 1.5 y 2.5, tiene menos estudiantes con GPA bajo y más estudiantes con GPA alto

La distribución de la desviación en estudiantes asiáticos es similar a las anteriores, pero con menos frecuencia, se nota un mayor número de estudiantes con GPA bajo comparado a los otros gráficos, de igual manera tiene una menor concentración de estudiantes con GPA alto.

Aproximadamente el 63% de los GPAs de las mujeres se encuentran a una distancia de una desviación estándar alrededor de la media. En una distribución normal ideal, se espera que el 68% de los datos estén dentro de 1σ.En este caso se tiene un 63%, lo cual es ligeramente menor, lo que sugiere que hay una leve dispersión o más valores alejados del promedio.

Casi el 98% de los GPAs de las mujeres están dentro de dos desviaciones estándar del promedio. Esto quiere decir que prácticamente todos los datos caen en este rango, confirmando que los valores extremos son muy pocos.
