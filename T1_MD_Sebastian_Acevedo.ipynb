##11. Determinar si la distribuci칩n de GPA sigue una distribuci칩n normal en los grupos genero, etnicidad y tutor칤a.

**Instrucciones:**
- Obten la desviaci칩n estandar del 'GPA' de las mujeres.
- Obten la desviaci칩n estandar del 'GPA' de los que reciben tutorias.
- Obten la desviaci칩n estandar del 'GPA' de los de etnicidad asi치tica.


- Genera un histograma que denote la distribuci칩n del 'GPA' de las mujeres.
- Genera un histograma que denote la distribuci칩n del 'GPA' de los que reciben tutorias.
- Genera un histograma que denote la distribuci칩n del 'GPA' de los de etnicidad asi치tica.


- Obten el porcentaje de datos dentro de 1픢 (1 * desviaci칩n estandar) para los datos de las mujeres.
- Obten el porcentaje de datos dentro de 2픢 (2 * desviaci칩n_estandar) para los datos de las mujeres.
- Concluye si la distribuci칩n de datos  de 'GPA' es normal para las mujeres a trav칠s de lo obtenido previamente.

```python
# Mujeres
dsv_mujeres = df[df['Gender'] == 1]['GPA'].std()
print(f"Desviaci칩n est치ndar del GPA (mujeres): {dsv_mujeres:.2f}")

# Con tutor칤as
dsv_tutoria = df[df['Tutoring'] == 1]['GPA'].std()
print(f"Desviaci칩n est치ndar del GPA (tutor칤a): {dsv_tutoria:.2f}")

# Etnicidad asi치tica
dsv_asia = df[df['Ethnicity'] == 2]['GPA'].std()
print(f"Desviaci칩n est치ndar del GPA (asi치ticos): {dsv_asia:.2f}")

# Histograma Mujeres
plt.figure(figsize=(6,6))
plt.hist(df[df['Gender'] == 1]['GPA'], bins=15, color='orchid', edgecolor='black')
plt.title('Distribuci칩n de GPA - Mujeres')
plt.xlabel('GPA')
plt.ylabel('Frecuencia')
plt.show()

# Histograma Tutor칤a
plt.figure(figsize=(6,6))
plt.hist(df[df['Tutoring'] == 1]['GPA'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribuci칩n de GPA - Con Tutor칤a')
plt.xlabel('GPA')
plt.ylabel('Frecuencia')
plt.show()

# Histograma Asi치ticos
plt.figure(figsize=(6,6))
plt.hist(df[df['Ethnicity'] == 2]['GPA'], bins=15, color='lightgreen', edgecolor='black')
plt.title('Distribuci칩n de GPA - Etnicidad Asi치tica')
plt.xlabel('GPA')
plt.ylabel('Frecuencia')
plt.show()

# GPA de mujeres
gpa_mujeres = df[df['Gender'] == 1]['GPA']
media = gpa_mujeres.mean()
dsv = gpa_mujeres.std()

# 1 desviaci칩n est치ndar
un_sigma = gpa_mujeres[(gpa_mujeres >= media - dsv) & (gpa_mujeres <= media + dsv)]
porcentaje_un_sigma = len(un_sigma) / len(gpa_mujeres) * 100
print(f"Porcentaje dentro de 1픢 (mujeres): {porcentaje_un_sigma:.2f}%")

# 2 desviaciones est치ndar
dos_sigma = gpa_mujeres[(gpa_mujeres >= media - 2*dsv) & (gpa_mujeres <= media + 2*dsv)]
porcentaje_dos_sigma = len(dos_sigma) / len(gpa_mujeres) * 100
print(f"Porcentaje dentro de 2픢 (mujeres): {porcentaje_dos_sigma:.2f}%")
```

La distribuci칩n de la desviaci칩n en las mujeres es bastante sim칠trica y cercana a la media, con un mayor n칰mero de estudiantes concentrados en los rangos de GPA de 1.5 a 2.5.
